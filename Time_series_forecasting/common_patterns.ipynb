### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_series(time, series, format = "-", start = 0, end = None, label = None):
    plt.plot(time[start:end], series[start:end], series[start:end], format, label = label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize = 14)
    plt.grid(True)
```

### Trend and Seasonality

```python
def trend(time, slope = 0):
    return time * slope
```

```python
time = np.arange(4 * 365 + 1)
baseline = 10
series = baseline + (time * .1)

plt.figure(figsize= (8, 5))
plot_series(time, series)
plt.show()
```

```python
print(time)
print(series)
```

Generate a time series.

```python
def seasonal_pattern(season_time):
    return np.where(season_time < 0.4,
                   np.cos(season_time * 2 * np.pi),
                   1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude, phase = 0):
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
amplitude = 40
series = seasonality(time, period = 730, amplitude = amplitude, phase = 100)
print(series)
plt.figure(figsize = (8, 5))
plot_series(time, series)
plt.show()
```

Time series with both trend and seasonality.

```python
slope = 0.05
series = baseline + trend(time, slope) + seasonality(time, period = 365, amplitude = amplitude)

plt.figure(figsize = (10, 6))
plot_series(time, series)
plt.show()
```

### Noise

```python
def white_noise(time, noise_level = 1, seed = None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
import seaborn as sns
noise_level = 5
noise = white_noise(time, noise_level, 42)

sns.histplot(noise, kde = True, bins = 20, color = 'r')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plot_series(time, noise)
plt.show()
```

Add white noise to the time series

```python
series += noise

plt.figure(figsize = (10, 6))
plot_series(time, series)
plt.show()
```
