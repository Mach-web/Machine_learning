### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_series(time, series, format = "-", start = 0, end = None, label = None):
    plt.plot(time[start:end], series[start:end], series[start:end], format, label = label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize = 14)
    plt.grid(True)
```

### Trend and Seasonality

```python
def trend(time, slope = 0):
    return time * slope
```

```python
time = np.arange(4 * 365 + 1)
baseline = 10
series = baseline + (time * .1)

plt.figure(figsize= (8, 5))
plot_series(time, series)
plt.show()
```

Generate a time series.

```python
def seasonal_pattern(season_time):
    return np.where(season_time < 0.4,
                   np.cos(season_time * 2 * np.pi),
                   1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude, phase = 0):
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
amplitude = 40
series = seasonality(time, period = 730, amplitude = amplitude, phase = 100)
print(series)
plt.figure(figsize = (8, 5))
plot_series(time, series)
plt.show()
```
